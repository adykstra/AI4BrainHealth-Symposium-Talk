## The mismatch negativity, or MMN
#### Change-related component of the auditory evoked response elicited by any discriminable violation of acoustic regularity.
<br>
<center>
    <img src="./media/mmn.png" style="float: center;" alt="sql" width="800">
</center>

_Näätänen et al 2001, 2011_

## The mismatch negativity, or MMN
- **_Automatic, requiring neither attention nor consciousness._**
    - Can be observed while participants are distracted (e.g. by reading a book).
    - Can be observed in sleep.
    - Can be observed under anesthesia.
    - Can be observed in disorders of consciousness.
<br>

- We know, however, that the MMN is both:
    - Very sensitive to stimulus context and perceptual organization.
    - Highly predictive of imminent recovery in disorders of consciousness.
<br>

- We sought to examine the pre-conscious notion of the MMN in awake, healthy participants.
<br>

_Näätänen et al 2011; Chennu and Bekinschtein 2012; Sussman et al 2014; Morlet and Fischer 2014_

## Informational masking

- Importantly:
    1. **_Even unperceived target tones reach auditory cortex._**
        - Spectral “protected region” prevents energetic masking.
        - Target tones well above sensory threshold.
    2. Target-tone streams not always present.
        - Permits false-alarm rates and d’
    3. Target-stream pitch varies across trials (but constant within a trial).
        - Target uncertainty
